```python
import pandas as pd

df1 = pd.read_csv("E:\\Synthetic-Email-Generation-tool-main\\data\\final_training_data1.csv")
df2 = pd.read_csv("E:\\Synthetic-Email-Generation-tool-main\\Candidate_application\\candidate_application_labeled.csv")

combined_df = pd.concat([df1, df2], ignore_index=True)

combined_df.to_csv("E:\\Synthetic-Email-Generation-tool-main\\data\\combined_with_candidate_application.csv", index=False)

print("CSV files merged successfully!")
```
